```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('/content/Walmart_Sales.csv')

print("First 5 rows of the dataset:")
df.head()
```

```python

df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format="%d-%m-%Y")
```

```python

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()

df[['Date', 'Year', 'Month']].head()
```

```python
df_2010 = df[df['Year'] == 2010]

print("Total records in 2010:", df_2010.shape[0])
```

```python
mean_weekly_sales = df_2010['Weekly_Sales'].mean()
median_weekly_sales = df_2010['Weekly_Sales'].median()
std_weekly_sales = df_2010['Weekly_Sales'].std()
var_weekly_sales = df_2010['Weekly_Sales'].var()

print("DESCRIPTIVE STATISTICS FOR WEEKLY SALES (2010)")
print("Mean Weekly Sales:", mean_weekly_sales)
print("Median Weekly Sales:", median_weekly_sales)
print("Standard Deviation:", std_weekly_sales)
print("Variance:", var_weekly_sales)
```

```python
monthly_sales = df_2010.groupby('Month')['Weekly_Sales'].sum()

print("MONTH-WISE TOTAL WEEKLY SALES (2010)")
monthly_sales
```

```python

plt.figure()
monthly_sales.plot(kind='bar')
plt.title('Monthly Sales â€“ 2010')
plt.xlabel('Month')
plt.ylabel('Total Weekly Sales')
plt.show()
```

```python

highest_sales_month = monthly_sales.idxmax()
highest_sales_value = monthly_sales.max()

print("HIGHEST WEEKLY SALES MONTH (2010)")
print("Month:", highest_sales_month)
print("Weekly Sales:", highest_sales_value)
```
