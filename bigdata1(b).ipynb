```python

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Walmart_Sales.csv')

df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()

df_2011 = df[df['Year'] == 2011]

print("Total records in 2011:", df_2011.shape[0])

monthly_total_sales = df_2011.groupby('Month')['Weekly_Sales'].sum()

print("\nMONTH-WISE TOTAL WEEKLY SALES (2011)")
print(monthly_total_sales)

highest_sales_month = monthly_total_sales.idxmax()
highest_sales_value = monthly_total_sales.max()

print("\nHIGHEST WEEKLY SALES MONTH (2011)")
print("Month:", highest_sales_month)
print("Total Weekly Sales:", highest_sales_value)

monthly_avg_sales = df_2011.groupby('Month')['Weekly_Sales'].mean()

print("\nMONTH-WISE AVERAGE WEEKLY SALES (2011)")
print(monthly_avg_sales)

plt.figure()
monthly_avg_sales.plot(kind='bar')
plt.title('Average Weekly Sales per Month – 2011')
plt.xlabel('Month')
plt.ylabel('Average Weekly Sales')
plt.show()

best_month_data = df_2011[df_2011['Month'] == highest_sales_month]

print("\nDIAGNOSTIC SUMMARY FOR HIGHEST WEEKLY SALES MONTH")
print(best_month_data[['Weekly_Sales']].describe())

monthly_sales_count = df_2011.groupby('Month')['Weekly_Sales'].count()

print("\nMONTH-WISE WEEKLY SALES RECORD COUNT (2011)")
print(monthly_sales_count)

plt.figure()
monthly_sales_count.plot(kind='bar')
plt.title('Number of Weekly Sales Records per Month – 2011')
plt.xlabel('Month')
plt.ylabel('Weekly Sales Record Count')
plt.show()
```
